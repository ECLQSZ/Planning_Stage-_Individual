# Questions

####  Broad question： What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?

#### Specific question: Can the variables in players.csv (experience, played_hours) be used to predict subscribe (TRUE/FALSE)?

#### During the data preparation stage, I will only use the players.csv file. I may convert the experience variable into a numerical format, such as assigning a value of 1 to “beginner”. Both this and played_hours will serve as explanatory variable X, while subscribe acts as response variable y(categorical). Before applying the KNN classifier, I will tune the K and standardize both x to ensure they are on a comparable scale. Then use KNN classification to predict and analyze players' subscription tendencies across different feature combinations. The target player's subscription status (TRUE/FALSE) will be predicted based on the majority category among the K nearest neighbors.


# Exploratory Data Analysis and Visualization

 ### Demonstrate that the dataset can be loaded into R.

```r
library(tidyverse)
players <- read_csv("data/players.csv")
sessions <- read_csv("data/sessions.csv")
```

### Minimum necessary wrangling

```r
players_clean <- players |>
  select(experience, subscribe, played_hours) |>
   mutate(experience_numerical_form = recode(experience,
                           "Beginner" = 1,
                           "Amateur" = 2, 
                           "Veteran" = 3,
                           "Pro" = 4,
                            .default = NA_real_)) |>
  filter(!is.na(played_hours) & !is.na(experience_numerical_form))  
players_clean
```

#### Compute the mean value for each quantitative variable in the players.csv data set

```r
players_summary <- players |>
  summarise(
    mean_played_hours = mean(played_hours, na.rm = TRUE),
    mean_age = mean(Age, na.rm = TRUE))
players_summary
```

### Visualizations

```r
p1 <- players_clean |>
  ggplot(aes(x = subscribe)) +
  geom_bar(fill = "steelblue", alpha = 0.7) +
  labs(
    title = "Distribution of subscribe",
    x = "Ture or False",
    y = "Number of Players"
  ) +
  theme_minimal()

p1
```

```r
p2 <- players_clean |>
  ggplot(aes(x = experience, fill = subscribe)) +
  geom_bar(position = "fill")  +
  labs(
    title = "Proportion of subscribe in different level of experience ",
    x = "Experience Level",
    y = "Proportion",
    fill = "Subscribe"
  ) +
  theme_minimal()

p2
```

## Methods and Plan

#### The K-Nearest Neighbors (KNN) method is appropriate because the research objective is to predict a categorical binary outcome based on a player's played hours and experience level. KNN assumes that samples at close distances exhibit similarity. However, the KNN model also has certain limitations, such as the selection of the K value. Model selection will be based on tuning the value of K using cross-validation within the training data. The dataset will be split into a training set (70%) and a testing set (30%).
